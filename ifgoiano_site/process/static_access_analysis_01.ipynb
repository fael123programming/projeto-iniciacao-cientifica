# Leitura do dataset mais recente

```python
import pandas as pd
import os
from datetime import datetime
import util


pd.set_option('display.float_format', lambda x: '%.2f' % x)

data = util.read_latest_dataset()
```

# Quantidade de publicações agrupadas pelo número de imagens

```python
data[['titulo', 'imagens']].groupby('imagens').count().reset_index().sort_values('titulo')
```

```python
from matplotlib import pyplot

df1 = data[['titulo', 'imagens']].groupby('imagens').count().reset_index().sort_values('titulo')
pyplot.rcParams['figure.figsize'] = (18, 10)
pyplot.title('Quantidade de publicações agrupadas pelo número de imagens')
pyplot.rcParams.update({'font.size': 30})
pyplot.xlabel('Quantidade de imagens')
pyplot.ylabel('Quantidade de publicações')
pyplot.bar(df1['imagens'], df1['titulo'])
```

# Soma da quantidade total de acessos agrupando as publicações pelo número de imagens

```python
data[['acessos', 'imagens']].groupby('imagens').sum().reset_index().sort_values('acessos')
```

```python
df2 = data[['acessos', 'imagens']].groupby('imagens').sum().reset_index().sort_values('acessos')
pyplot.ticklabel_format(style='plain')
pyplot.title('Soma da quantidade total de acessos agrupando as publicações pelo número de imagens')
pyplot.rcParams.update({'font.size': 30})
pyplot.xlabel('Quantidade de imagens')
pyplot.ylabel('Soma total de acessos')
pyplot.bar(df2['imagens'], df2['acessos'])
```

# Quantidade de publicações por dia da semana e período do dia
